```python
from typing import Self
card_value_dict = {
    'skullking':17,
    'pirate': 16,
    'mermaid': 15,
}
class Card(str):
    color: str
    number: int

    def __new__(cls, value: str) -> Self:
        obj = str.__new__(cls, value)
        color, *number = obj.split('-')
        print(number)
        number = ''.join(number)
        print(number)
        obj.color = color
        obj.number = int(number) if number else (card_value_dict[color] if color in card_value_dict else 0)
        print(obj, obj.number)
        return obj

    def __init__(self, *args, **kwargs) -> None:
        self.color = self.color
        self.number = self.number

    def __repr__(self) -> str:
        if self.endswith('-'):
            return self[:-1]
        return super().__repr__()
cards = [Card('yellow-5'), Card('black-14'), Card('black-12'), Card('black-13'), Card('green-13'), Card('yellow-13'), Card('yellow-6')]
cards[0].number
```
